You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('CSV_files/epa-sea-level.csv')
```

```python
df.head()
```

```python
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'])
plt.xlabel('Years')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.show()
```

```python
model= linregress(x=df.Year, y=df['CSIRO Adjusted Sea Level'])
```

```python
slope = model.slope
```

```python
intercept= model.intercept
```

```python
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'])
plt.plot(df.Year, intercept + slope*df.Year, c='r')
plt.xlabel('Years')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
year = np.arange(1880, 2051)
```

```python
predicted_df = pd.DataFrame({'Year':year})
```

```python
predicted_df.head()
```

```python
predicted_df['Estimated Adjusted Sea Level'] =predicted_df.Year.apply(lambda x: slope*x+intercept)
```

```python
predicted_df.head()
```

```python
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'])
plt.plot(predicted_df.Year, predicted_df['Estimated Adjusted Sea Level'], c='r')

# labels
plt.xlabel('Years')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
# Create scatter plot
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'])

# regression 
model= linregress(x=df.Year, y=df['CSIRO Adjusted Sea Level'])
slope = model.slope
intercept= model.intercept

# predicted dataframe
predicted_df = pd.DataFrame({'Year':np.arange(1880, 2051)})
predicted_df['Estimated Adjusted Sea Level'] =predicted_df.Year.apply(lambda x: slope * x + intercept)

# Create first line of best fit till 2050
plt.scatter(df.Year, df['CSIRO Adjusted Sea Level'])
plt.plot(predicted_df.Year, predicted_df['Estimated Adjusted Sea Level'])

# data from year 2000 
df_2000 = df.loc[df.Year >= 2000]

# linear model for 2000
model = linregress(df_2000.Year, df_2000['CSIRO Adjusted Sea Level'])
slope = model.slope
intercept = model.intercept

# prediction
predicted_df_2000 = predicted_df.loc[predicted_df.Year >= 2000]
predicted_df_2000.loc[: ,'Estimated Adjusted Sea Level'] =predicted_df_2000['Year'].apply(lambda x: slope*x+intercept)

# Create second line of best fit
plt.scatter(df_2000.Year, df_2000['CSIRO Adjusted Sea Level'])
plt.plot(predicted_df_2000.Year, predicted_df_2000['Estimated Adjusted Sea Level'])

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```
